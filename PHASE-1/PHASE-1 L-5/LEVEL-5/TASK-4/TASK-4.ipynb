```python
import pandas as pd
import matplotlib.pyplot as plt
from dataclasses import dataclass

@dataclass
class DatasetStats:
    mean: float
    max: float
    min: float

    def plot_histogram(self, column, df):
        """Plot a histogram of a specified column."""
        plt.hist(df[column], bins=20, edgecolor='black')
        plt.xlabel(column)
        plt.ylabel("Frequency")
        plt.title(f"Histogram of {column}")
        plt.show()

df = pd.read_csv("housing.csv")

stats = DatasetStats(
    mean=df["price"].mean(),
    max=df["price"].max(),
    min=df["price"].min()
)

while True:
    choice = input("\nType 'stats' to view stats, 'plot' to see a histogram, or 'exit' to quit: ").strip().lower()

    if choice == "stats":
        print(f"\nMean: {stats.mean:.2f}, Max: {stats.max}, Min: {stats.min}")
    
    elif choice == "plot":
        column = input("\nEnter column name to plot histogram (e.g., 'price'): ").strip().lower()
        if column in df.columns:
            stats.plot_histogram(column, df)
        else:
            print("\nInvalid column name. Try again.")
    
    elif choice == "exit":
        print("\nExiting.")
        break
    
    else:
        print("\nInvalid input. Try again.")

```
