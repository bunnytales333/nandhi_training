```python
import numpy as np
import matplotlib.pyplot as plt
n = 20     
p = 0.5    
size = 10000  
data = np.random.binomial(n, p, size)
mean_simulated = np.mean(data)
plt.figure(figsize=(8, 5))
plt.hist(data, bins=np.arange(n+2)-0.5, density=True, alpha=0.7, color='purple', edgecolor='green', label="Simulated Data")
plt.axvline(mean_simulated, color='pink', linestyle='dashed', linewidth=2, label=f"Mean: {mean_simulated:.2f}")
plt.xlabel("Number of Successes")
plt.ylabel("Probability Density")
plt.title("Binomial Distribution Simulation")
plt.xticks(range(n+1))
plt.legend()
plt.show()
```
