```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    "id": [1, 2, 3, 4, 5],
    "price": [300000, 450000, 200000, 600000, 800000],
    "sqft_living": [1500, 2000, 1000, 2500, 3000],
    "bedrooms": [3, 4, 2, 5, 6],
    "bathrooms": [2, 3, 1, 4, 5]
}
df = pd.DataFrame(data)
df["price_per_sqft"] = df["price"] / df["sqft_living"]
print("\nUpdated DataFrame with price_per_sqft:")
print(df)
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["sqft_living"], y=df["price_per_sqft"], hue=df["bedrooms"], palette="viridis", size=df["bathrooms"])
plt.title("Price per Sqft vs. Living Area")
plt.xlabel("Sqft Living")
plt.ylabel("Price per Sqft")
plt.legend(title="Bedrooms", bbox_to_anchor=(1, 1))
plt.show()
```
